### 문제 2. 지뢰찾기 문제

```python
import random
def create_mine(n):
    tmp_list = [['' for i in range(n)] for k in range(n)]
    special_chr = ['.','*']
    for i in range(n):
        for k in range(n):
            tmp_list[i][k] = special_chr[random.randint(0,1)]
            print(tmp_list[i][k],end='  ')
        print('\n')
    return tmp_list

mine_list = create_mine(9)
```

```python
## 패딩을 사용해서 풀이해야 함
def print_mine_result(tmp_list):
    n = len(tmp_list)
    for i in range(n):
        for k in range(n):
            if tmp_list[i][k] == '*':
                print('*',end='  ')
            else:
                sum = 0
                ## 좌상단
                if i==0 and k==0:
                    if tmp_list[i+1][k] == '*':
                        sum += 1
                    if tmp_list[i][k+1] == '*':
                        sum += 1
                    if tmp_list[i+1][k+1] == '*':
                        sum += 1
                ## 우상단
                elif i==0 and k==n-1:
                    if tmp_list[i+1][k] == '*':
                        sum += 1
                    if tmp_list[i][k-1] == '*':
                        sum += 1
                    if tmp_list[i+1][k-1] == '*':
                        sum += 1
                ## 좌하단
                elif i==n-1 and k==0:
                    if tmp_list[i-1][k] == '*':
                        sum += 1
                    if tmp_list[i][k+1] == '*':
                        sum += 1
                    if tmp_list[i-1][k+1] == '*':
                        sum += 1
                ## 우하단
                elif i==n-1 and k==n-1:
                    if tmp_list[i-1][k] == '*':
                        sum += 1
                    if tmp_list[i][k-1] == '*':
                        sum += 1
                    if tmp_list[i-1][k-1] == '*':
                        sum += 1
                ## 좌변
                elif 0<i<n-1 and k==0:
                    if tmp_list[i-1][k] == '*':
                        sum += 1
                    if tmp_list[i-1][k+1] == '*':
                        sum += 1
                    if tmp_list[i][k+1] == '*':
                        sum += 1
                    if tmp_list[i+1][k+1] == '*':
                        sum += 1
                    if tmp_list[i+1][k] == '*':
                        sum += 1
                ## 우변
                elif 0<i<n-1 and k==n-1:
                    if tmp_list[i-1][k] == '*':
                        sum += 1
                    if tmp_list[i-1][k-1] == '*':
                        sum += 1
                    if tmp_list[i][k-1] == '*':
                        sum += 1
                    if tmp_list[i+1][k-1] == '*':
                        sum += 1
                    if tmp_list[i+1][k] == '*':
                        sum += 1
                ## 상변
                elif i==0 and 0<k<n-1:
                    if tmp_list[i][k-1] == '*':
                        sum += 1
                    if tmp_list[i+1][k-1] == '*':
                        sum += 1
                    if tmp_list[i+1][k] == '*':
                        sum += 1
                    if tmp_list[i+1][k+1] == '*':
                        sum += 1
                    if tmp_list[i][k+1] == '*':
                        sum += 1
                ## 하변
                elif i==n-1 and 0<k<n-1:
                    if tmp_list[i][k-1] == '*':
                        sum += 1
                    if tmp_list[i-1][k-1] == '*':
                        sum += 1
                    if tmp_list[i-1][k] == '*':
                        sum += 1
                    if tmp_list[i-1][k+1] == '*':
                        sum += 1
                    if tmp_list[i][k+1] == '*':
                        sum += 1
                ## 나머지
                else:
                    if tmp_list[i-1][k-1] == '*':
                        sum += 1
                    if tmp_list[i-1][k] == '*':
                        sum += 1
                    if tmp_list[i-1][k+1] == '*':
                        sum += 1
                    if tmp_list[i][k-1] == '*':
                        sum += 1
                    if tmp_list[i][k+1] == '*':
                        sum += 1
                    if tmp_list[i+1][k-1] == '*':
                        sum += 1
                    if tmp_list[i+1][k] == '*':
                        sum += 1
                    if tmp_list[i+1][k+1] == '*':
                        sum += 1
                    
                print(sum,end='  ')
                
        print('\n')
    
print_mine_result(mine_list)
```

```python
## 패딩을 사용한 방식
def print_mine_result2(tmp_list):
    n = len(tmp_list)
    
    pad=[]
    for i in range(n+2):
        pad.append('.')
        
    
    for i in range(n):
        tmp_list[i].insert(0,'.')
        tmp_list[i].append('.')
    tmp_list.insert(0,pad)
    tmp_list.append(pad)
        
    for i in range(1,n+1):
        for k in range(1,n+1):
            if tmp_list[i][k] == '*':
                print('*',end='  ')
            else:
                sum = 0
                for l in range(i-1,i+2):
                    for m in range(k-1,k+2):
                        if tmp_list[l][m] == '*':
                            sum += 1
                print(sum,end='  ')
        print('\n')
                
print_mine_result2(mine_list)
```

### 문제 3. tanh(x)함수의 그래프 그리기

```python
import matplotlib.pyplot as plt
# %matplotlib inline
import matplotlib as mpl
import matplotlib.font_manager as fm
import numpy as np
import pandas as pd
import math

# MacOS 한글 설정
from matplotlib import rc
rc('font', family='AppleGothic')
mpl.rcParams['axes.unicode_minus'] = False   # minus 표시

# Windows 한글 설정
# from matplotlib import rc
# rc('font', family='D2Coding')
# mpl.rcParams['axes.unicode_minus'] = False   # minus 표시

font2 = {'size':18}

xdata = np.arange(-2*math.pi,2*math.pi, 0.01)
ydata = np.tanh(xdata)
    
plt.title('y = tanh',fontdict=font2)
plt.plot(xdata, ydata, linewidth=3)
plt.xlim(-2*math.pi,2*math.pi)
plt.ylim(-1.2,1.2)

plt.grid(True)
plt.show()
```

### 문제 4. 멜론 top100 크롤링

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import math
from tqdm.notebook import tqdm
import requests

RANK = 100 ## 멜론 차트 순위가 1 ~ 100위까지 있음

header = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Trident/7.0; rv:11.0) like Gecko'}
req = requests.get('https://www.melon.com/chart/month/index.htm', headers = header) ## 주간 차트를 크롤링 할 것임
html = req.text
parse = BeautifulSoup(html, 'html.parser')

titles = parse.find_all("div", {"class": "ellipsis rank01"}) 
singers = parse.find_all("div", {"class": "ellipsis rank02"}) 
albums = parse.find_all("div",{"class": "ellipsis rank03"})

title = []
singer = []
album = []

for t in titles:
    title.append(t.find('a').text)

for s in singers:
    singer.append(s.find('span', {"class": "checkEllipsis"}).text)

for a in albums:
    album.append(a.find('a').text)
```

```python
before_ranks = parse.find_all("span", {"class": "rank_wrap"}) 
before_ranks
```

```python
before_ranks[0].find('span',{"class":"none"}).get_text()
```

```python
before_ranks[4].find('span',{"class":"none"}).get_text()
before_ranks[4].find('span',{"class":"down"}).get_text()
```

```python
before_rank = []
rank = 1
for br in before_ranks:
    if br.find('span',{"class":"none"}).get_text() == '순위 진입':
        before_rank.append(-1)
    elif br.find('span',{"class":"none"}).get_text() == '순위 동일':
        before_rank.append(rank)
    elif br.find('span',{"class":"none"}).get_text() == '단계 하락':
        num = int(br.find('span',{"class":"down"}).get_text())
        before_rank.append(rank+num)
    else:
        num = int(br.find('span',{"class":"up"}).get_text())
        before_rank.append(rank-num)
    rank += 1
```

```python
df = pd.DataFrame({'rank':range(1,101),'before_rank':before_rank,'title':title,'singer':singer,'album':album})
df.set_index('rank',inplace=True)
df
```

```python
print('현재순위\t전월순위\t가수 : 제목(앨범)')
for i in range(10):
    if df['before_rank'].iloc[i] == -1:
        print('%3d\tnew\t%s : [ %s ] - %s'
              %((i+1),
                df['singer'].iloc[i],
                df['title'].iloc[i],
                df['album'].iloc[i]))
    else:
        print('%3d\t%3d\t%s : [ %s ] - %s'
              %((i+1),
                df['before_rank'].iloc[i],
                df['singer'].iloc[i],
                df['title'].iloc[i],
                df['album'].iloc[i]))
```
