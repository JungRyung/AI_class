```python
# ex 12-3
font1 = {'size':18,'color':'green'}

xdata = ['안지영', '홍지수', '황예린']
ydata = [90,85,88]
plt.plot(xdata,ydata)

plt.title('세명 학생의 국어 성적',fontdict=font1)
plt.xlabel('이름')
plt.ylabel('성적')

plt.show()
```

```python
# ex 12-4 선 그래프의 범례
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
plt.plot(xdata,ydata1, label='국어')
plt.plot(xdata,ydata2, label='영어')
plt.legend(loc='upper center')

plt.title('세명 학생의 국어, 영어 성적',fontdict=font1)

plt.show()
```

```python
# ex 12-5 그래프 선 스타일 설정
xdata = ['안지영', '홍지수', '황예린']
ydata1 = [90,85,88]
ydata2 = [83,88,91]
ydata3 = [85,97,78]
ydata4 = [92,88,82]

plt.plot(xdata,ydata1, label='국어', color='red', linestyle='-', marker='o', linewidth=3, markersize=10)
plt.plot(xdata,ydata2, label='영어', color='orange', linestyle='--', marker='*', linewidth=3, markersize=10)
plt.plot(xdata,ydata3, label='수학', color='green', linestyle='-.', marker='s', linewidth=3, markersize=10)
plt.plot(xdata,ydata4, label='사회', color='blue', linestyle=':', marker='d', linewidth=3, markersize=10)

plt.title('세명 학생의 네 과목 성적')
plt.legend(loc='best')

plt.show()
```

```python
# ex 12-6 X/Y축의 범위와 눈금 설정
font2 = {'size':18}

xdata = list(range(30))
ydata = []

for x in xdata:
    y = 2 * x
    ydata.append(y)
    
plt.title('y = 2x',fontdict=font2)
plt.plot(xdata, ydata, linewidth=3)
plt.xlim(0,30)
plt.ylim(0,60)
plt.xticks(list(range(-1,30,2)))
plt.yticks(list(range(-2,60,5)))

plt.grid(True)
plt.show()
```

```python
# y = sinx 그리기
import numpy as np
import math

font2 = {'size':18}

xdata = np.arange(0,2*math.pi, 0.01)
ydata = np.sin(xdata)
    
plt.title('y = sinx',fontdict=font2)
plt.plot(xdata, ydata, linewidth=3)
plt.xlim(0,2*math.pi)
plt.ylim(-1.2,1.2)

plt.grid(True)
plt.show()
```

```python
# y = sinx 그리기
import numpy as np
import math

font2 = {'size':18}

xdata = np.arange(0.1,10, 0.01)
ydata = np.log2(xdata)
    
plt.title('y = log2x',fontdict=font2)
plt.plot(xdata, ydata, linewidth=3)
plt.xlim(-1,10)
plt.ylim(-5,5)

plt.grid(True)
plt.show()
```
