# 네이버 영화 평점 크롤링 (에어로너츠)

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import math
```

```python
url_base = "https://movie.naver.com"
url_sub = "/movie/bi/mi/pointWriteFormList.nhn?code=167658&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page=1"

url = url_base + url_sub

html = urlopen(url)
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
## iframe의 src 주소값 얻기
ifr_url = soup.find('iframe')['src']
ifr_url
```

```python
## 평점 개수 구하기
total_num = int(soup.find_all(class_='total')[0].find('em').get_text())
total_num
```

```python
## 페이지 수 구하기
page_num = math.ceil(total_num/10)
page_num
```

```python
uls = soup.find_all(class_='score_result')
rest_list = uls[0].find_all('li')
tmpOne = rest_list[0]
tmpOne
```

```python
star_score = tmpOne.find(class_='star_score').get_text().replace('\n','')
star_score
```

```python
reple = tmpOne.find(id='_filtered_ment_0').get_text().strip()
reple
```

```python
date = tmpOne.find(class_='score_reple').find('dt').find_all('em')[1].get_text()
date
```

```python
star_score = []
reple = []
date = []

num = 0
for rest in rest_list:
        star_score.append(rest.find(class_='star_score').get_text().replace('\n',''))
        reple.append(rest.find(id='_filtered_ment_'+str(num)).get_text().strip())
        date.append(rest.find(class_='score_reple').find('dt').find_all('em')[1].get_text())
        num +=1
```

```python
df = pd.DataFrame({'star_score':star_score,'reple':reple,'date':date})
df
```

```python
url_front = "https://movie.naver.com/movie/bi/mi/pointWriteFormList.nhn?code=167658&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page="

star_score = []
reple = []
date = []

for page in range(1,page_num+1):
    url = url_front+str(page)
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')
    uls = soup.find_all(class_='score_result')
    rest_list = uls[0].find_all('li')
    num = 0
    for rest in rest_list:
        star_score.append(rest.find(class_='star_score').get_text().replace('\n',''))
        reple.append(rest.find(id='_filtered_ment_'+str(num)).get_text().strip())
        date.append(rest.find(class_='score_reple').find('dt').find_all('em')[1].get_text())
        num += 1        
```

```python
df = pd.DataFrame({'star_score':star_score,'reple':reple,'date':date})
df['date'] = pd.to_datetime(df['date'])
df.sort_values(by='date', ascending = True, inplace=True)
df
```

```python
df.to_csv('movie_grade.csv',sep=',',encoding="utf-8")
```
