## 30개 찾기

```python
url_front = "https://www.menupan.com/restaurant/search/search_main_sub_area.asp?page="
url_back = "&areacode=dj202&trec=67"

cafe_name = []
scores = []
kinds = []
addr = []
tel = []

for page in range(1,4):
    url = url_front+str(page)+url_back
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')
    uls = soup.find_all(class_='areaList')
    rest_list = uls[0].find_all(class_='ifo')
    
    for rest in rest_list:
        cafe_name.append(rest.find('a').get_text())
        scores.append(rest.find(class_='total').get_text())
        kinds.append(rest.find_all('dd')[1].get_text().split('|')[0])
        addr.append(rest.find_all('dd')[1].get_text().split('|')[1])
        tel.append(rest.find_all('dd')[2].get_text())
```

```python
df = pd.DataFrame({'cafe':cafe_name,'score':scores,'kind':kinds,'addr':addr,'tel':tel})
df
```

```python
df.to_csv('gungdong1.csv',sep=',',encoding="utf-8")
```

```python
import numpy as np
import folium
import googlemaps
from tqdm.notebook import tqdm
import time
```

```python
key_fd = open('../googlemapskey.txt', mode='r')
gmaps_key = key_fd.read(100)
key_fd.close()
```

```python
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
lat = []
lng = []

for n in tqdm(df.index):
    if df['addr'][n] != 'Multipl':
        target_name = df['addr'][n]+', '+'Chicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['lat'] = lat
df['lng'] = lng
df.head()
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],zoom_start=11)
folium.Marker([df['lat'].mean(), df['lng'].mean()],popup='center').add_to(mapping)
mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()],zoom_start=11)

for n in df.index:
    if df['addr'][n] != 'Multipl':
        folium.Marker([df['lat'][n], df['lng'][n]],popup=df['cafe'][n]).add_to(mapping)
mapping
```
