```python
# Q13-3
df = pd.read_csv('high_school_2019.csv', skiprows=[0], names=['1학년 학급수','1학년 학생수','2학년 학급수','2학년 학생수','3학년 학급수','3학년 학생수'])
df
```

```python
df.iloc[2:] = df.iloc[2:].astype(int)
```

```python
df['총학생수'] = df[['1학년 학생수','2학년 학생수','3학년 학생수']].sum(axis=1)
df
```

```python
# Q13-4
df2 = df[['1학년 학생수','2학년 학생수','3학년 학생수']].sum()
df2

for i in range(3):
    print('%d학년 총 학생수 : %d'%(i,df2[i]))
```

```python
df2.plot.bar(title='전국 고등학교 학년별 총 학생 수')
plt.show()
```
