## 여자 친구 선물 고르기

```python
from bs4 import BeautifulSoup 
from urllib.request import urlopen
import urllib
import time
```

```python
# https://search.naver.com/search.naver?where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%EC%97%AC%EC%B9%9C+%EC%84%A0%EB%AC%BC&c_id=&c_name=&sm=tab_pge&kin_start=2
tmp1 = 'https://search.naver.com/search.naver?'
html = tmp1 + 'where=kin&kin_display=10&qt=&title=0&&answer=0&grade=0&choice=0&sec=0&nso=so%3Ar%2Ca%3Aall%2Cp%3Aall&query=%EC%97%AC%EC%B9%9C+%EC%84%A0%EB%AC%BC&c_id=&c_name=&sm=tab_pge&kin_start={num}'

response = urlopen(html.format(num=1))
soup = BeautifulSoup(response, "html.parser")
tmp = soup.find_all('dl')
```

```python
tmp_list = []
for line in tmp:
    tmp_list.append(line.text)
    
tmp_list[:5]
```

```python
from tqdm import tqdm_notebook

present_candi_text = []
for n in tqdm_notebook(range(1, 1000, 10)):
    response = urlopen(html.format(num=n))
    soup = BeautifulSoup(response, "html.parser")
    tmp = soup.find_all('dl')
    for line in tmp:
        line_str = line.text.replace('\t','').replace('\n','')
        present_candi_text.append(line.text)
        
    time.sleep(0.5)
```

```python
with open("data/21_present_candi.txt", "w",-1,'utf-8') as f:
    for line in present_candi_text:
        f.write(line+'\n')
```

```python
present_text = ''

for each_line in present_candi_text:
    present_text = present_text + each_line + '\n'
```
